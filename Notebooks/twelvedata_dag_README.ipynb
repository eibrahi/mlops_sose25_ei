# 📊 Airflow DAG: `stock_data_collector_twelvedata`

Automatisiert den **stündlichen Abruf von Intraday-Aktienkursen** über die TwelveData API. Die Kursdaten werden pro Aktie gespeichert und bei neuen Zeitpunkten inkrementell erweitert.

## 🚀 Funktionsweise
Dieser DAG implementiert folgende Schritte:
1. **Initialisiert Konfigurationen und API-Key**
2. **Prüft lokale CSV-Dateien je Aktie auf den letzten Zeitstempel**
3. **Ruft neue Daten ab**, sofern verfügbar (unter Beachtung des 30-Tage-Limits im Free-Plan)
4. **Speichert neue Daten** je Aktie als CSV im Verzeichnis `stock_data`

## ⚙️ Setup-Voraussetzungen
- Airflow Umgebung mit Zugriff auf:
  - TwelveData API Key (`TWELVEDATA_API_KEY` als Umgebungsvariable)
  - Lokales Datenverzeichnis `/home/holu/airflow/data/stock_data`
- Python Libraries:
  - `pandas`, `requests`, `pytz`, `random`, `logging`

## 🕒 Zeitplan
- **Schedule:** `0 * * * *` → jede volle Stunde
- **Ausführung:** Serienmäßig, d. h. Aktie für Aktie nacheinander

## 🧠 Features & Hinweise
- **User-Agent-Rotation** zur Minimierung von Blockierungen
- **Timezone-Handling** mit UTC → Europe/Berlin-Konvertierung
- **Logging pro Aktie**, auch bei API-Fehlern oder leeren Rückgaben
- **Fallback-Logik** für neu erstellte Dateien

## 📁 Dateiablage
CSV-Dateien werden gespeichert unter:
`/home/holu/airflow/data/stock_data/<TICKER>.csv`

Beispiel:
```bash
stock_data/
├── AAPL.csv
├── MSFT.csv
└── TSLA.csv
```

## ✅ Best Practices umgesetzt
- `catchup=False` → nur Echtzeit-Ausführungen
- `max_active_runs=1` → keine Parallelisierung
- `retry_delay` & `retries` eingebaut für Stabilität
- `dag.test()` für manuelles Testen im Local-Setup
