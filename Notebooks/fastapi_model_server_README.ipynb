# 🧠 FastAPI Modellserver mit MLflow

Dieser FastAPI-Service stellt **Aktienkursprognosen** bereit, basierend auf einem MLflow-Modell. Er lädt automatisch Modell und Features beim Start, überwacht Änderungen an der CSV-Datei und bietet zwei REST-Endpoints:

## 🔌 Endpunkte

### `GET /health`
Healthcheck für Modellstatus, Featureverfügbarkeit und CSV-Integrität.
- Gibt ✅ / ❌ Statusmeldungen zurück
- Liefert `503`, wenn etwas fehlt oder fehlschlägt

### `GET /api/v1/predictions`
Gibt pro Aktie folgende Vorhersage zurück:
- Letzter Schlusskurs (`last_close`)
- Prognostizierte Rendite (`predicted_return`)
- Vorhergesagter nächster Preis (`predicted_price`)
- Letzter Timestamp der zugrundeliegenden Daten

## ⚙️ Architekturüberblick
- **Modell:** geladen aus MLflow Registry (Stage `Production`)
- **Features:** dynamisch geladen aus `info.json`
- **CSV:** `/data/stock_data/combined_stock_data.csv` mit Ticker-Daten
- **Auto-Reload:** CSV wird bei Dateiänderung neu geladen

## 📂 Datei- & Projektstruktur
```bash
data/
├── stock_data/
│   └── combined_stock_data.csv
├── models/
│   └── info.json  # Feature Columns aus MLflow
└── fastapi_model_server.py
```

## 🚀 Beispielstart (lokal)
```bash
uvicorn fastapi_model_server:app --reload --port 8000
```

## 🔐 Anforderungen
- MLflow Tracking Server (`http://localhost:5001`)
- Modell muss im MLflow unter `stock_price_model` registriert sein
- Kombinierte CSV muss vorhanden sein und Spalte `stock`, `Close`, etc. enthalten

## ✅ Zusammenfassung der Features

| Bereich | Beschreibung |
|---|---|
| **Modellverwaltung** | Lädt Produktionsmodell aus MLflow inkl. `info.json` |
| **CORS-Handling** | Universell geöffnet für API-Zugriffe (auch lokal) |
| **Healthcheck** | Zeigt Status von Modell, Features und CSV-Datei |
| **/predictions** | Pro Aktie: Rückgabe von `last_close`, `predicted_return`, `predicted_price` |
| **Auto-Reload CSV** | Lädt CSV neu, falls Datei geändert wurde |
