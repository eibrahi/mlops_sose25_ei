# 🚀 Airflow DAG: `serve_model_api`

Dieser DAG dient dem automatisierten **Neustart eines FastAPI-Servers**, der ein MLflow-Modell bereitstellt. Damit ist sichergestellt, dass der Server regelmäßig neu gestartet wird und aktuelle Modellversionen verwendet.

## 🔧 Funktion
Der DAG führt regelmäßig folgende Aktionen aus:
1. Prüft, ob bereits ein `uvicorn`-Serverprozess läuft
2. Falls ja, wird der Prozess beendet (via `psutil`)
3. Startet den FastAPI-Server mit dem Skript `fastapi_model_server.py`
4. Führt den Server im Entwicklungsmodus mit `--reload` aus

## 📂 Struktur & Pfade
- `DATA_DIR`: Verzeichnis mit Stockdaten
- `MODEL_DIR`: Speicherort für MLflow-Modelldaten
- `FASTAPI_SCRIPT`: Pfad zur API-Anwendung, z. B. `/home/holu/airflow/data/fastapi_model_server.py`

## 🕒 Zeitplan
| Einstellung        | Wert             |
|---------------------|------------------|
| `schedule_interval` | `15 * * * *`     |
| `catchup`           | `False`          |
| `retry_delay`       | `5 Minuten`      |
| `retries`           | `1`              |

## 🛡️ Hinweise zur Robustheit
- Beendet vorhandene `uvicorn`-Prozesse, um Port-Konflikte zu vermeiden
- Nutzt `cwd=os.path.dirname(...)` zur Sicherstellung korrekter Modulpfade
- **Hot Reload aktiv:** nützlich zur Entwicklung, aber ggf. für Produktion deaktivieren

## 📈 Erweiterungsideen
- 🔍 Healthcheck-Task mit HTTP-Request auf `/health`
- 📬 Slack/Webhook-Benachrichtigung bei Startproblemen
- 🧪 Automatisierte Tests zur Serververfügbarkeit

## ✅ Voraussetzungen
- Python Packages: `fastapi`, `uvicorn`, `psutil`, `subprocess`, `airflow`
- FastAPI-Modellserver (siehe `fastapi_model_server.py`)
- MLflow Tracking Server sollte bereits laufen

Startbefehl für MLflow Beispiel:
```bash
mlflow server \
  --backend-store-uri ./mlruns \
  --default-artifact-root ./artifacts \
  --host 0.0.0.0 \
  --port 5001
```
